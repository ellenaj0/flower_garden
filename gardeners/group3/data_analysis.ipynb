```python
# Since all total growth values are the same anyways we can just use the mean (or max or min would be the same)
df_filtered.groupby(['Gardener', 'Config', 'Turn'])[['Total_Growth']].mean().reset_index()
```

### Total growth over all turns for all gardeners and configurations

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import re


df_mean = df_filtered.groupby(['Gardener', 'Config', 'Turn'])['Total_Growth'].mean().reset_index()

# Sort according to config num
df_mean[['config_num1', 'config_num2']] = df_mean['Config'].str.extract(r'config(\d+)_(\d+)').astype(int)
df_mean = df_mean.sort_values(['config_num1', 'config_num2']).reset_index(drop=True)
df_mean = df_mean.drop(columns=['config_num1', 'config_num2'])

configs = df_mean['Config'].unique()
print(f"Unique configurations: {len(configs)}")

colors = cm.get_cmap('tab20')

for i, config in enumerate(configs):
    print(f"Config: {config} ({i+1}/{len(configs)})")

    plt.figure(figsize=(7, 4))
    df_config = df_mean[df_mean['Config'] == config]

    gardeners = df_config['Gardener'].unique()

    for j, gardener in enumerate(gardeners):
        df_g = df_config[df_config['Gardener'] == gardener]
        plt.plot(df_g['Turn'], df_g['Total_Growth'],
                 label=gardener,
                 color=colors(j / len(gardeners)))

    plt.title(f"Config: {config}")
    plt.xlabel("Turn")
    plt.ylabel("Total Growth")
    plt.legend(title="Gardener", bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()
```
